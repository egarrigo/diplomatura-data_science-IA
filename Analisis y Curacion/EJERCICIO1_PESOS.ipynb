## Ejercicio 1

Graficar Curvas correspondientes a cada percentil de peso para niños y niñas
* La columna SEX = 1 agrupa a los niños y SEX = 2 a las niñas
*Las curvas de los percentiles correspondietes a un mismo sexo deben ir en un mismo grafico
*Un grafico para cada sexo

```python
%matplotlib inline
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import seaborn
```

```python
dataset = pd.read_csv('wtageinf.csv')
```

```python
dataset
```

```python
dataset.dtypes
```

```python
dataset.columns
```

```python
filtrado1= dataset[dataset.Sex==1]
filtrado1
```

```python
filtrado2= dataset[dataset.Sex==2]
filtrado2
```

```python
y = ['P3', 'P5', 'P10', 'P25', 'P50', 'P75', 'P90', 'P95']
```

```python
plt.figure(figsize=(18,6))
plt.subplot(121)

for peso in y:
    seaborn.lineplot(x='Agemos', y= peso, data = filtrado1, ci = None)#, legend = False)
plt.title('Curva de Peso de NIÑOS en funcion de la edad en Percentiles')    
plt.ylabel('Peso en Kg')
plt.xlabel('Edad en Meses')
plt.legend(y[:])    

    
plt.subplot(122)
for peso in y:
    seaborn.lineplot(x='Agemos', y= peso, data = filtrado2, ci = None)#, legend = False)

plt.title('Curva de Peso de NIÑAS en funcion de la edad en Percentiles')    
plt.ylabel('Peso en Kg')
plt.xlabel('Edad en Meses')
plt.legend(y[:])
```

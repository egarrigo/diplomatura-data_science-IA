*Hacer un grafico en barras agrupando por valor del titulo "province"

```python
import pandas as pd
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
pfront = pd.read_json ('pasos.json')
```

```python
pfront
```

```python
pprov= pfront.groupby(['province']).count().reset_index()
pprov
```

```python
plt.figure(figsize=(18,6))
plt.bar(pprov['province'],pprov['location'])
plt.title('Cantidad de Pasos fronterizos de cada Provincia de Argentina')    
plt.ylabel('Cantidad')
plt.xlabel('Provincias')
plt.xticks(rotation=45)
plt.show()
```
